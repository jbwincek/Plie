```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('speed_test_log.csv')
```

```python
df.columns = ['not_join()', 'using_join_poorly()', 'join_all_at_once()']
#df['run'] = pd.Series(np.arange(0,df.shape[0]), index=df.index)
```

```python
current_palette = sns.color_palette('pastel')
sns.set_palette(current_palette)
sns.set_style('white')


plt.plot(df)
plt.xlabel('runs')
plt.ylabel('execution time')
plt.legend(['not_join()', 'using_join_poorly()', 'join_all_at_once()'])
plt.show()
```

```python
print(df.shape)
print('not_join() median runtime {}'.format(np.median(df['not_join()'])))
print('using_join_poorly() median runtime {}'.format(np.median(df['using_join_poorly()'])))
print('join_all_at_once() median runtime {}'.format(np.median(df['join_all_at_once()'])))
```
